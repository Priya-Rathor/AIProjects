```python
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt
```

```python
model_id1 = "dreamlike-art/dreamlike-diffusion-1.0"
model_id2 = "stabilityai/stable-deffusion-xl-base-1.0"
pipe1 = StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float16,use_safetensors = True)
pipe1 = pipe1.to("cuda")
```

```python
prompt =""" dreamlikeart, a grungy woman with rainbow hair, travelling between dimensions, dynamic pose ,happy,soft eyse and narrow chain,
extreme bokeh, dainty figure, long hair straight down , torn kawaii shirt and  baggy jeans
"""
```

```python
image = pipe1(prompt).images[0]
```

```python
image
```

```python
print("[PROMPT]:",prompt)
plt.imshow(image);
plt.axis('off')
```
